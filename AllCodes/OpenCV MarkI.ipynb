```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
plnt = cv2.imread("planets.jpg")
```

```python
plnt.shape
```

Matplotlib RGB formatında renk sıralması yapmaktadır. OpenCV ise BGR formatında sıralama kullanmaktadır

```python
plt.imshow(plnt)
```

Bunun önüne geçmek için ise

```python
fixedimage = cv2.cvtColor(plnt, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fixedimage)
```

```python
cv2.rectangle(fixedimage, pt1 = (425, 10), pt2 = (565, 130), color = (255, 0, 0), thickness = 3)
```

```python
plt.imshow(fixedimage)
```

```python
cv2.circle(fixedimage,center = (55,52), radius = 50, color = (0, 200, 0), thickness = 3)
plt.imshow(fixedimage)
```
