Eval fonksiyonu kendisine veriler karakter dizisi değerlendirip bir sonuç olarak döndürür

```python
avengers = cv2.imread("avengers.jpg")
avengers = cv2.cvtColor(avengers, cv2.COLOR_BGR2RGB)

height, width, channel = avengers.shape

tony = cv2.imread("avengerstony.jpg")
tony = cv2.cvtColor(tony, cv2.COLOR_BGR2RGB)

methods = ["cv2.TM_CCOEFF", "cv2.TM_CCOEFF_NORMED", "cv2.TM_CCORR", "cv2.TM_CCORR_NORMED", "cv2.TM_SQDIFF", "cv2.TM_SQDIFF_NORMED"]
```

```python
for m in methods:
    avengerscopy = avengers.copy()
    
    method = eval(m)
    
    res = cv2.matchTemplate(avengerscopy, tony, method)
    
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
        
    bottom_right = (top_left[0] + width, top_left[1] + height)
    
    cv2.rectangle(avengerscopy, top_left, bottom_right, (0,255,0), 4)
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title("Isı haritasında belirleme")
    
    plt.subplot(122)
    plt.imshow(avengerscopy)
    plt.title("Detection İşlemi")
    
    plt.suptitle(m)
    plt.show()
    print('\n')
    print('\n')
```
