```python
import cv2
import matplotlib.pyplot as plt
```

```python
imbgr = cv2.imread("snn.png")
spacebgr = cv2.imread("images.png")
```

```python
im = cv2.cvtColor(imbgr, cv2.COLOR_BGR2RGB)
space = cv2.cvtColor(spacebgr, cv2.COLOR_BGR2RGB)
```

```python
im.shape
```

```python
space.shape
```

```python
space = cv2.resize(space, (500,500))
```

```python
space.shape
```

```python
blended = cv2.addWeighted(src1 = im, alpha = 0.5, src2 = space, beta = 0.5, gamma = 0)
```

```python
plt.imshow(blended)
```
